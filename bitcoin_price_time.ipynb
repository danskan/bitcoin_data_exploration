```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#load file 
file_to_load = 'bitcoin_data.csv'

#read file and load into pandas dataframe
price_data = pd.read_csv(file_to_load)
price_data.head()
```

```python
#re-order rows based on ascending data
price_data = price_data.sort_values(by=['End'])
price_data.head(150)
```

```python
plt.yscale('log')
plt.scatter(price_data.End, price_data.Close)
plt.show()
```

```python
plt.plot(price_data['End'],price_data['Close'])
plt.xticks(price_data['End'])
plt.xlabel("Daily Close")
plt.ylabel('Price')
plt.title('Price vs Date')
plt.yscale('log')
plt.xscale('log')
plt.show()

#need to explore data sources...
```
