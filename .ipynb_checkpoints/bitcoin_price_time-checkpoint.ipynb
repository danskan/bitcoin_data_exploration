```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#load file 
file_to_load = 'bitcoin_data.csv'

#read file and load into pandas dataframe
price_data = pd.read_csv(file_to_load)
price_data.head()
```

```python
#re-order rows based on ascending data
price_data = price_data.sort_values(by=['End'])
```

```python
# inspect and cleaning underlying datatypes
price_data.info()
price_data['End'] = pd.to_datetime(price_data['End'])
price_data['Start'] =pd.to_datetime(price_data['Start'])
```

```python
plt.yscale('log')
plt.xscale('log')
plt.scatter(price_data.End, price_data.Close)
plt.show()
```

```python
plt.plot(price_data['End'],price_data['Close'])
plt.xticks(price_data['End'])
plt.xlabel("Daily Close")
plt.ylabel('Price')
plt.title('Price vs Date')
plt.yscale('log')
plt.xscale('log')
plt.show()

#need to explore data sources...
```

```python
# adding line of regression using the OLS (ordinary least squares model)
import statsmodels.api as sm

r_data = pd.read_csv('bitcoin_data.csv')

# add column that includes days since genesis block

# define variables
x = r_data['Close'].tolist()
y = r_data['days'].tolist()

# add constant term
x = sm.add_constant(x)

# perform regression
result = sm.OLS(y, x).fit()
print(result.summary())
```
